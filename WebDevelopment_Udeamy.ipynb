```python
data_descending['price_2'] = np.where(data_descending['isPaid']== True, data_descending['price'], 0)
```

```python
data_descending['price'].unique()
```

```python
data_descending.price.value_counts().plot(kind='bar')
```

```python
data_price = data_descending.groupby(['price'],as_index=False).sum()
```

```python
data_price.sort_values(by='numSubscribers', ascending=False)
```

```python
data_price.plot(x='price', y = 'numSubscribers', kind='bar', figsize=(10,5))
plt.ylabel('Number of Subscribers')
plt.xlabel('price in dollars')
```

```python
data_price.plot(x='price', y = 'numReviews', kind='bar', figsize=(10,5))
plt.ylabel('Number of reviewers')
plt.xlabel('price in dollars')
```
